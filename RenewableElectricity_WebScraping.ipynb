```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_renewable_electricity_production'
site = requests.get(url)
soup = BeautifulSoup(site.text, 'html')
```

## Table

Table I want to scrape

```python
table = soup.find_all('table')[2] #chose the table i wanted to scrape
```

```python
print(table)
```

## Column Name

Just checking the column names (for verification)

```python
title = table.find_all("th")
```

```python
table_title = [title.text.strip() for title in title]
```

```python
print(table_title)
```

```python
df = pd.DataFrame(columns = table_title)

df
```

# Data Frame

Converting the content I scraped from Wikipedia into a DataFrame. Which, I can later on, download as csv.

```python
column_data = table.find_all('tr')
print(column_data) 


```

```python
for row in column_data[1:]:
    row_data = row.find_all(['td', 'th'])
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
